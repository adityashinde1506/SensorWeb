```python
import numpy
import matplotlib.pyplot as plotter
from scipy.signal import *
```

```python
indexes = numpy.linspace(start=0, stop=100, num=100)
noise = numpy.random.normal(size=indexes.shape)
test_signal = numpy.clip(a=square(2 * numpy.pi * 10* indexes, duty=0.5), a_min=0, a_max=1) + (noise * 0.1)
```

```python
plotter.plot(indexes,test_signal)
plotter.show()
```
