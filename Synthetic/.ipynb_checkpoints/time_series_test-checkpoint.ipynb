```python
import numpy
import matplotlib.pyplot as plotter
```

## Make sine wave

### Dataset 1

```python
sine_a1 = numpy.array([numpy.sin(i*0.01) + numpy.random.normal(0,1) for i in range(10000)])
plotter.plot(sine_a1)
plotter.show()
```

```python
sine_a2 = numpy.array([numpy.sin(i*0.05) + numpy.random.normal(0,0.1) for i in range(10000)])
plotter.plot(sine_a2)
plotter.show()
```

```python
data1 = numpy.vstack([sine_a1, sine_a2])
data1.shape
```

### Dataset 2

```python
sine_b1 = numpy.array([numpy.sin(i*0.001) + numpy.random.normal(0,1) for i in range(10000)])
plotter.plot(sine_b1)
plotter.show()
```

```python
sine_b2 = numpy.array([numpy.sin(i*0.002) for i in range(10000)])
plotter.plot(sine_b2)
plotter.show()
```

```python
data2 = numpy.vstack([sine_b1, sine_b2])
data2.shape
```

### Plot both datasets

```python
plotter.scatter(data1[0,:], data1[0,:], c='r', marker='+', alpha=0.5)
plotter.show()
plotter.scatter(data2[0,:], data2[0,:], c='b', marker='^', alpha=0.5)
plotter.show()
```

## Prepare time series data

```python
data1.shape
```

```python
data2.shape
```

```python
data1 = data1.T
data2 = data2.T
print(data1[:10])
print(data2[:5])
```

```python
data1 = data1.reshape((-1,5,2))
data2 = data2.reshape((-1,5,2))
print(data1[:5])
print(data2[:5])
```

```python
batch_size = data1.shape[0]
data1 = data1.reshape((batch_size,-1))
data2 = data2.reshape((batch_size,-1))
print(data1.shape)
print(data2.shape)
print(data1[:5])
print(data2[:5])
```

```python
pca = PCA(n_components=2)
pca.fit(numpy.vstack([data1, data2]))
```

```python
pdata1 = pca.transform(data1)
pdata2 = pca.transform(data2)
```

```python
plotter.scatter(pdata1[:,0], pdata1[:,1], c='r', marker='+', alpha=0.5)
plotter.scatter(pdata2[:,0], pdata2[:,1], c='b', marker='^', alpha=0.5)
plotter.show()
```

## Try classification

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.neural_network import MLPClassifier
```

```python
data1.shape
```

```python
data2.shape
```

```python
data1y = numpy.zeros(data1.shape[0])
data2y = numpy.zeros(data2.shape[0]) + 1
print(data1y)
print(data2y)
```

```python
y = numpy.hstack([data1y, data2y])
X = numpy.vstack([data1, data2])
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr = LogisticRegression(verbose=1, max_iter=100, solver='sag')
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
sgdc = SGDClassifier(alpha=0.1, verbose=0, max_iter=10000, loss='perceptron')
sgdc.fit(X_train, y_train)
sgdc.score(X_test, y_test)
```

```python
mlpc = MLPClassifier(hidden_layer_sizes=(100,50), solver='adam', verbose=1)
mlpc.fit(X_train, y_train)
mlpc.score(X_test, y_test)
```
